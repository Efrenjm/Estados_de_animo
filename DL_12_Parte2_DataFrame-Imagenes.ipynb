## Objetivo:
    
    
En esta lección se mostrará como hacer una base de datos basado en imagenes.
   
 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2 as cv
from glob import glob
from sklearn.utils import shuffle
```

## Generación del dataframe de testeo

### Generación del dataframe "negativo"

```python
count = 0
df_negative_test = pd.DataFrame()
for imagen in glob('datos/faces_test/negatives/*.jpg'):
    img = cv.imread(imagen)
    img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    img_r = cv.resize(img_gray, (40,40), interpolation = cv.INTER_NEAREST)
    fila = np.append(0, img_r.flatten())
    fila = fila.reshape(1, 1601  )
    df_negative_test = pd.concat([df_negative_test,  pd.DataFrame(fila)  ])
    count = count + 1
print(count)
df_negative_test
```

### Generación del dataframe "positivo"

```python
count = 0
df_postive_test = pd.DataFrame()
for imagen in glob('datos/faces_test/positives/*.jpg'):
    img = cv.imread(imagen)
    img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    img_r = cv.resize(img_gray, (40,40), interpolation = cv.INTER_NEAREST)
    fila = np.append(1, img_r.flatten())
    fila = fila.reshape(1, 1601  )
    df_postive_test = pd.concat([df_postive_test,  pd.DataFrame(fila)  ])
    count = count + 1
print(count)
df_postive_test
```

### Unión de dataframes 

Cración de un solo dataframe llamado df_faces y se arreglan de manera aleatoria

```python
frames = [df_postive_test, df_negative_test]
df_faces_test = pd.concat(frames)
df_faces_test = shuffle(df_faces_test)
df_faces_test
```

## Generación del dataframe de aprendizaje

### Generación del dataframe "negativo"

### Generación del dataframe "positivo"

```python
count = 0
df_postive = pd.DataFrame()
for imagen in glob('datos/faces/positives/*.jpg'):
    img = cv.imread(imagen)
    img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    img_r = cv.resize(img_gray, (40,40), interpolation = cv.INTER_NEAREST)
    fila = np.append(1, img_r.flatten())
    fila = fila.reshape(1, 1601  )
    df_postive = pd.concat([df_postive,  pd.DataFrame(fila)  ])
    count = count + 1
print(count)
```

### Unión de dataframes 

Cración de un solo dataframe llamado df_faces y se arreglan de manera aleatoria

```python
frames = [df_postive, df_negative]
df_faces = pd.concat(frames)
df_faces = shuffle(df_faces)
df_faces
```
